```python
import autogluon.core as ag
from autogluon.vision import ImagePredictor

# 90:10(Train : Test)
train_dataset, _, test_dataset = ImagePredictor.Dataset.from_folders('../content/drive/MyDrive/base_dir')
print(train_dataset)


model = ag.Categorical('resnet50_v1', 'mobilenetv3_small')
model_list = ImagePredictor.list_models()
batch_size = 32
lr = ag.Categorical(1e-2, 1e-3)
hyperparameters={'model': model, 'batch_size': batch_size, 'lr': lr, 'epochs': 50}

predictor = ImagePredictor()
# split from train dataset to train and validate (90:10)
predictor.fit(train_dataset, search_strategy='bayesopt', time_limit=60*720, hyperparameters = hyperparameters)
print('Top-1 val acc: %.3f' % predictor.fit_summary()['valid_acc'])

fit_result = predictor.fit_summary()
print('Top-1 train acc: %.3f, val acc: %.3f' %(fit_result['train_acc'], fit_result['valid_acc']))

image_path = test_dataset.iloc[0]['image']
result = predictor.predict(image_path)
print(result)

proba = predictor.predict_proba(image_path)
print(proba)

bulk_result = predictor.predict(test_dataset)
print(bulk_result)

image_path = test_dataset.iloc[0]['image']
feature = predictor.predict_feature(image_path)
print(feature)

top1, top5 = predictor.evaluate(test_dataset)
print('Top-1 test acc: %.3f' % top1)
print('Top-5 test acc: %.3f' % top5)


filename = 'model.ag'
predictor.save(filename)
predictor_loaded = ImagePredictor.load(filename)
result = predictor_loaded.predict(image_path)
print(result)
```
